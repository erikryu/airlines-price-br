```python
# Bibliotecas utilizadas

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import pandas as pd
from pandas.io import gbq
```

```python
# Passo 1


# Abrir um navegador

pasta = ""
navegador = webdriver.Firefox()

# Abrir o site da Latam

# O local de partida dos voos é Brasília

navegador.get("https://www.latamairlines.com/br/pt/destinos")

# Apertar botão de cookies 

WebDriverWait(navegador, 10).until(EC.element_to_be_clickable((By.XPATH,'//*[@id="cookies-politics-button"]'))).click()


```

```python
# Apertar no botão de mais ofertas

while True:
    try:
        WebDriverWait(navegador, 10).until(EC.element_to_be_clickable((By.XPATH, '//*[@id = "btnPagging"]'))).click()
        print("Botão com mais opções clicado")
    except TimeoutException:
        break
```

```python
# Criando vetores para as variáveis 

destino       = []
data_ida      = []
tipo_passagem = []
tipo_classe   = []
tipo_voo      = []
preco         = []
```

```python
# Coletando destinos

destinos_nav = navegador.find_elements(By.XPATH, '//*[@class = "sc-dZoequ iRXMAL latam-typography latam-typography--heading-05 sc-eZkCL miRdz"]')

# Adicionando valores ao vetor

for i in destinos_nav:
    destino.append(i.text)
    
# Coletando dia de ida

data_ida_nav = navegador.find_elements(By.XPATH, '//*[@class = "sc-dZoequ bKICOT latam-typography latam-typography--paragraph-medium sc-eZkCL miRdz"]//strong')

# Adicionando valores ao vetor

for i in data_ida_nav:
    data_ida.append(i.text)

# Coletando tipo de passagem

tipo_passagem_nav = navegador.find_elements(By.XPATH, '//*[@class = "sc-dZoequ jcENnf latam-typography latam-typography--paragraph-medium sc-eZkCL miRdz"]')

# Adicionando valores ao vetor

for i in tipo_passagem_nav:
    tipo_passagem.append(i.text)

# Coletando tipo de classe

tipo_classe_nav = navegador.find_elements(By.XPATH, '//*[@class = "sc-dZoequ jvTWQM latam-typography latam-typography--paragraph-medium sc-eZkCL miRdz"]')

# Adicionando valores ao vetor

for i in tipo_classe_nav:
    tipo_classe.append(i.text)
    
# Coletando tipo de voo

tipo_voo_nav = navegador.find_elements(By.XPATH, '//div[@class = "sc-kdIgRK dWlbGe"]')

# Adicionando valores ao vetor

for i in tipo_voo_nav:
    tipo_voo.append(i.text)
    
# Coletando preço

preco_nav = navegador.find_elements(By.XPATH, '//*[@class = "sc-dZoequ hsWRFd latam-typography latam-typography--heading-05 sc-eZkCL miRdz"]')

# Adicionando valores ao vetor

for i in preco_nav:
    preco.append(i.text)
```

```python
# Sair do navegador 

navegador.quit()
```

```python
# Passo 2

voos = {
    'Destino': destino,
    'ida': data_ida,
    'passagem': tipo_passagem,
    'classe': tipo_classe,
    'voo': tipo_voo,
    'preco': preco
}
# Criando data frame

dados_voos = pd.DataFrame(voos)

# Apresentando dados

display(dados_voos)
```

```python
#tratamento da coluna 'voo'
dados_voos['voo'] = dados_voos['voo'].str.strip()
display(dados_voos)
```

```python
# Passo 3

# As variáveis tipo_voo e preço precisam de tratamento 

# Tipo de voo

# Separando variáveis taxa e tipo de voo
dados_voos[["tipo_taxa", "voo"]] = dados_voos["voo"].str.split('-', n=1, expand=True)

# Preço

# Separando preço em duas colunas

dados_voos[["moeda", "preco"]] = dados_voos['preco'].str.split(' ', n=1, expand=True)

# Transformando a coluna preço para numérico

dados_voos['preco'] = dados_voos['preco'].str.replace(".", "")
dados_voos['preco'] = dados_voos['preco'].str.replace(",", ".")

dados_voos['preco'] = pd.to_numeric(dados_voos['preco']).astype(float)
```
